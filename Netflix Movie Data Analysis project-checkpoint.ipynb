# categories vote_ Average column

We would cut the vote_Average values and make 4 categoreis popular avrage below_avg not_popular to describe it more using catigorize_col() function
provided above.

```python
def categorize_col(df, col, labels):

    edge = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]

    df[col] = pd.cut(df[col], bins=edge, labels=labels, duplicates='drop')
    
    return df
```

```python
labels = ['not_popular', 'below_avg', 'average', 'popular']

categorize_col(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```
